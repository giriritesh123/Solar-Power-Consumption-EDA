```python
import matplotlib.pyplot as plt

# Total consumption across all zones
df[['Zone 1 Power Consumption', 'Zone 2  Power Consumption', 'Zone 3  Power Consumption']].plot(figsize=(12,6))

plt.title("Power Consumption Trend (All Zones)")
plt.xlabel("Time Index")
plt.ylabel("Consumption (kW)")
plt.legend(["Zone 1", "Zone 2", "Zone 3"])
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['Zone 1 Power Consumption'], label='Zone 1', color='blue')
plt.plot(df['Zone 2  Power Consumption'], label='Zone 2', color='orange')
plt.plot(df['Zone 3  Power Consumption'], label='Zone 3', color='green')

plt.title("Power Consumption Trend by Zone")
plt.xlabel("Time Index")
plt.ylabel("Consumption (kW)")
plt.legend()
plt.show()
```
